## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A) = svv_19 = Valor_Total_Vendas
tkt_19 = 320.52
sqv_19 = 3
svv_19 = tkt_19 * sqv_19
print(svv_19)
```

```
# (B) = sqv_20 = Qtd_Total_Vendas
svv_20 = 834.47
tkt_20 = 119.21
sqv_20 = svv_20 / tkt_20
print(sqv_20)
```

```
# (C) = tkt_23 = Ticket Medio
svv_23 = 15378.12
sqv_23 = 5
tkt_23 = svv_23 / sqv_23
print(tkt_23)
```

```
tkt = (tkt_19 + tkt_20 + tkt_23) / 3
print(tkt)
```
